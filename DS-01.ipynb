```python
#6 Find Measure of Central Tendency: MEAN, MEDIAN, MODE for Degree, Inter and
print("Mean of degree =",data['DEGREE'].mean())
print("Median degree =",data['DEGREE'].median())
print("Mode degree =",data['DEGREE'].mode())
```

```python
print("Mean inter =",data['INTERMEDIATE'].mean())
print("Median inter =",data['INTERMEDIATE'].median())
print("Mode inter =",data['INTERMEDIATE'].mode())
```

```python
print("Mean 10th =",data['SSC'].mean())
print("Median 10th =",data['SSC'].median())
print("Mode 10th =",data['SSC'].mode())
```

```python
#7 Find Measure of Variance: Minimum, Maximum, Range, Mean Deviation, Standard Deviation,Co-efficient of Variation for Degree,Inter and 10th
print("Minimum =",data['DEGREE'].min())
print("Maximum =",data['DEGREE'].max())
print("Range =",9.53-5.8)
print("Mean Deviation =",data['DEGREE'].mad())
print("Standard Deviation =",data['DEGREE'].std())
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
print("Co-efficient of Variation =",cv(data['DEGREE']))
```

```python
print("Minimum =",data['INTERMEDIATE'].min())
print("Maximum =" ,data['INTERMEDIATE'].max())
print("Range =",99.4-65)
print("Mean Deviation =",data['INTERMEDIATE'].mad())
print("Standard Deviation =",data['INTERMEDIATE'].std())
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
print("Co-efficient of Variation =",cv(data['INTERMEDIATE']))
```

```python
print("Minimum =",data['SSC'].min())
print("Maximum =",data['SSC'].max())
print("Range =",99-38.4)
print("Mean Deviation =",data['SSC'].mad())
print("Standard Deviation =",data['SSC'].std())
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
print("Co-efficient of Variation =",cv(data['SSC']))
```

```python
#8 Measures of Position: Standard Scores, Inter-quartile Range for Degree, Inter and 10th
import numpy as np
import scipy.stats as stats
print(stats.zscore(data['DEGREE']))
print("-----------------------------------------")
q3 = np.percentile(data['DEGREE'],75)
q1 = np.percentile(data['DEGREE'],25)
iqr = q3 - q1
print('The Interquartile Range for DEGREE =',iqr)
```

```python
import numpy as np
import scipy.stats as stats
print(stats.zscore(data['INTERMEDIATE']))
print("-----------------------------------------")
q3 = np.percentile(data['INTERMEDIATE'],75)
q1 = np.percentile(data['INTERMEDIATE'],25)
iqr = q3 - q1
print('The Interquartile Range for INTERMEDIATE =',iqr)
```

```python
import numpy as np
import scipy.stats as stats
print(stats.zscore(data['SSC']))
print("-----------------------------------------")
q3 = np.percentile(data['SSC'],75)
q1 = np.percentile(data['SSC'],25)
iqr = q3 - q1
print('The Interquartile Range for SSC =',iqr)
```

```python
#9 Create Box Plot and Identify Outliers for Degree, Inter and 10th 
data['DEGREE'].plot(kind='box')
plt.show()
data['INTERMEDIATE'].plot(kind='box')
plt.show()
data['SSC'].plot(kind='box')
plt.show()
def outlier(a):
 q1 = np.quantile(a,0.25)
 q2 = np.quantile(a,0.75)
 m = np.median(a)
 iqr = q2-q1
 u_bound = q2+(1.5*iqr)
 l_bound = q1-(1.5*iqr)
 # print(iqr,u_bound,l_bound)
 # print("Inter Quartile Range:",iqr)
 outliers = a[(a<= l_bound)|(a>= u_bound)]
 print("outliers in boxplot:\n{}".format(outliers))
outlier(data['DEGREE'])
outlier(data['INTERMEDIATE'])
outlier(data['SSC'])
```
